```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import numpy as np
```

```python
img1 = cv2.imread('../DATA/dog_backpack.png')
img2 = cv2.imread('../DATA/watermark_no_copy.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2 =cv2.resize(img2,(600,600))
x_offset=934-600
y_offset=1401-600
# I want to put logo on top-left corner, So I create a ROI
roi = img1[y_offset:1401,x_offset:943] # BOTTOM RIGHT CORNER
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
plt.imshow(roi)
```

```python
mask = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(mask,cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(mask)
```

```python
plt.imshow(mask_inv,cmap='gray')
```

```python
background = np.full(img2.shape, 255, dtype=np.uint8)
mask_3_channel = cv2.bitwise_and(background, background, mask=mask_inv)
mask_3_channel.shape
```

```python
plt.imshow(mask_3_channel,cmap='gray')
```

```python
invert_mask_3_channel = cv2.bitwise_not(mask_3_channel)
```

```python
plt.imshow(invert_mask_3_channel,cmap='gray')
```

```python
img1_bg = cv2.bitwise_or(roi,roi,mask = mask_inv)
plt.imshow(img1_bg,cmap='gray')
```
